### May Trips Data
    Some underlying things to remember:
    * trip ids reset every day with XXXn format, i.e. there are 31 LYF1 trips in the data set for July, because there is one 1st trip for every day of the month. It follows companies. 
    * Data is supposed to be scrubbed to remove trip durations 1h < trip < 24h
    * Scooters are supposed to be evaluated on a monthly basis to see if each and every scooter was used at least 3x. Operators may be required to reduce fleet size, but only after notice and hearing.
